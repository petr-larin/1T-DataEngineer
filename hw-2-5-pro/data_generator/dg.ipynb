```python
from datetime import date, timedelta
from scipy.stats import norm

avg_sales = {   
    'shop_dns' : [20, 25, 30],
    'shop_mvideo' : [25, 30, 35],
    'shop_sitilink' : [30, 35, 40]   }

avg_plan = {   
    'shop_dns' : [600, 750, 900],
    'shop_mvideo' : [750, 900, 1050],
    'shop_sitilink' : [900, 1050, 1200]   }


def print_sales(dt, shop_name):
    avs = avg_sales[shop_name]
    for pid in range(3):
        print("('{}', {}, {}),".format(dt, pid, max(0, int(norm.rvs(avs[pid], scale=8)))))
        

def print_plan(d):
    for shop in avg_plan:
        avp = avg_plan[shop]
        for pid in range(3):
            print("({}, '{}', {}, '{}'),".format(pid, shop, max(0, int(norm.rvs(avp[pid], scale=10))), d))
```

```python
start_date = date(2023, 1, 1)
end_date = date(2023, 3, 1)
```

```python
d = start_date
while d < end_date:
    print_sales(d, 'shop_sitilink')
    d += timedelta(days=1)
```

```python
d = start_date
while d < end_date:
    print_plan(d)
    d += timedelta(days=31)
    d = d.replace(day=1)
```
