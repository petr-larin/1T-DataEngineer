```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dataset_0.csv")
```

```python
df
```

```python
df.corrwith(df['class']).abs().sort_values()
```

```python
plt.figure(figsize=(16, 7))
cmap = sns.diverging_palette(230, 20, as_cmap=True)

plt.title('Ковариационная матрица')
sns.heatmap(df.corr(), cmap=cmap, annot=False, square=True)

plt.show()
```

Видно, что признаки feature_4, feature_20, feature_13, feature_17, feature_3 не имеют значимой корреляции или влияния на другие переменные в датасете.

```python
df = pd.read_csv("dataset_1.csv")
```

```python
df
```

```python
df.corrwith(df.target)
```

```python
plt.figure(figsize=(5, 5))
cmap = sns.diverging_palette(230, 20, as_cmap=True)

plt.title('Ковариационная матрица')
sns.heatmap(df.corr(), cmap=cmap, annot=False, square=True)

plt.show()
```

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(data=df, x="feature_1", y="feature_2", hue="target")
plt.show()
```

```python
df["feature_3"] = (df.feature_1**2 + df.feature_2**2)**0.5
```

```python
df.corrwith(df.target)
```

Видно, что свойства по отдельности друг с другом и таргетом не коррелируют, но коррелирует длина вектора (f1, f2).

```python
plt.figure(figsize=(5, 5))
cmap = sns.diverging_palette(230, 20, as_cmap=True)

plt.title('Ковариационная матрица')
sns.heatmap(df.corr(), cmap=cmap, annot=False, square=True)

plt.show()
```
